```python
import numpy as np
import pandas as pd
import os
import shutil
import pydicom as dicom
import cv2
```

```python
#configuracion de los parametros
savepath = 'C:/Users/Francisco/Desktop/Noveno Ciclo/Informatica Industrial/PROYECTO/DATASET/Data_convertida_separada_RSNA'

#Direccion para la base de datos del estudio de Cohen.
rsna_imgpath = 'C:/Users/Francisco/Desktop/Noveno Ciclo/Informatica Industrial/PROYECTO/RSNA Pneumonia Detection Dataset Challenge/TODAS LAS IMAGENES' 
rsna_xlsxpath = 'C:/Users/Francisco/Desktop/Noveno Ciclo/Informatica Industrial/PROYECTO/RSNA Pneumonia Detection Dataset Challenge/stage_2_train_labels.xlsx'
```

```python
rsna_xlsx = pd.read_excel(rsna_xlsxpath)
```

```python
filename_label = {'Neumonia': [], 'Normal': []}
count = {'Neumonia': 0, 'Normal':0}

for index, row in rsna_xlsx.iterrows():   #Iterar en cada fila de la metadata.xlsx
    f = row['Target'] # toma todas las filas de la columna finding
    #print (f)
    if f == 1: #    #Si el valor encontrado en la columna "finding" es igual a "COVID-19"
        count['Neumonia'] += 1  #Incrementar en 1 el valor de la clase correspondiente
        entry = row['patientId']
        filename_label['Neumonia'].append(entry+'.dcm')
    elif f==0:
        count['Normal'] += 1  #Incrementar en 1 el valor de la clase correspondiente
        entry = row['patientId']
        filename_label['Normal'].append(entry+'.dcm')
print(count)
#print(filename_label)
```

```python
#rsna_imgpath
for im in os.scandir(rsna_imgpath):
    print(im.name.split)
    if im.name in filename_label['Neumonia']:
        ds=dicom.dcmread(os.path.join(rsna_imgpath,im.name))
        m_pix=ds.pixel_array
        imgname=im.name + '.jpg'
        cv2.imwrite(os.path.join(savepath,'NEUMONIA',imgname),m_pix)
    elif im.name in filename_label['Normal']:
        ds=dicom.dcmread(os.path.join(rsna_imgpath,im.name))
        m_pix=ds.pixel_array
        imgname=im.name + '.jpg'
        cv2.imwrite(os.path.join(savepath,'NORMAL',imgname),m_pix)
```
